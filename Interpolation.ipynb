```python
# Nearest neighbour interpolation

import cv2
import numpy as np
img = cv2.imread('mango.jfif')
near_img = cv2.resize(img,None, fx = 5, fy = 5, interpolation = cv2.INTER_NEAREST)
cv2.imshow(" ",near_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('mango.jfif')
bilinear_img = cv2.resize(img,None, fx = 5, fy = 5, interpolation = cv2.INTER_LINEAR)
cv2.imshow(" ",bilinear_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('mango.jfif')
bicubic_img = cv2.resize(img,None, fx = 5, fy = 5, interpolation = cv2.INTER_CUBIC)
cv2.imshow(" ",bicubic_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
res = cv2.resize(img, None, fx=0.2, fy=0.2, interpolation = cv2.INTER_CUBIC)

plt.figure(figsize=(15,12))

plt.subplot(121)
plt.imshow(img,cmap = 'gray')
plt.title('Original Image')

plt.subplot(122)
plt.imshow(res,cmap = 'gray')
plt.title('Downsampled Image')

plt.show()
```
