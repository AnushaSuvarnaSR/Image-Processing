```python
# image sharpen
from PIL import Image
from PIL import ImageFilter
import matplotlib.pyplot as plt
# load the image
my_image=Image.open('house.jpg')
# use sharpen function
sharp=my_image.filter(ImageFilter.SHARPEN)
# save the image
sharp.save('D:image_sharpen.jpg')
sharp.show()
plt.imshow(sharp)
plt.show()
```

```python
# image flip
import matplotlib.pyplot as plt
# load the image
img=Image.open('house.jpg')
plt.imshow(img)
plt.show()
# use flip function
flip=img.transpose(Image.FLIP_LEFT_RIGHT)

# save the image
flip.save('D:image_flip.jpg')
plt.imshow(flip)
plt.show()
```
